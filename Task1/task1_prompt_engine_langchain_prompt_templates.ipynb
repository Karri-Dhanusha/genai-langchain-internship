## ✅ Task 1: Replace Hardcoded Prompts

Convert static prompts into reusable PromptTemplate.

```python
# Task 1: PromptTemplate

template = "Explain {topic} in simple terms for beginners"

prompt_template = PromptTemplate(
    input_variables=["topic"],
    template=template
)

# Test
output = prompt_template.format(topic="Artificial Intelligence")
print(output)
```

## ✅ Task 2: Multi-Input Prompt System

Creating a dynamic prompt with multiple inputs.

```python
multi_template = "Explain {topic} for {audience} in a {tone} tone"

multi_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=multi_template
)

# Test Cases
print(multi_prompt.format(topic="AI", audience="beginner", tone="friendly"))
print(multi_prompt.format(topic="Python", audience="kids", tone="fun"))
print(multi_prompt.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## ✅ Task 3: Prompt Variations Engine
Different templates for different use cases.

```python
# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## ✅ Task 4: ChatPromptTemplate System

Role-based prompt generation.

```python
def get_chat_prompt(role, topic):
    role_map = {
        "teacher": "You are a helpful teacher.",
        "interviewer": "You are a strict interviewer.",
        "motivator": "You are a motivational speaker."
    }

    system_message = role_map.get(role, "You are an assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Talk about {topic}")
    ])

    messages = chat_template.format_messages(topic=topic)

    # Clean Output
    for msg in messages:
        print(f"{msg.type.upper()}: {msg.content}")

# Test
get_chat_prompt("teacher", "Neural Networks")
```

## ✅ Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone, strict=False):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        if strict:
            raise ValueError("Invalid audience")
        print("Invalid audience → defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        if strict:
            raise ValueError("Invalid tone")
        print("Invalid tone → defaulting to 'formal'")
        tone = "formal"

    return audience, tone
```

## ✅ Task 6: Prompt Generator App

```python
# Template Dictionary (Separation of Concerns)
TEMPLATES = {
    "teaching": "Explain {topic} clearly step by step for {audience}s in a {tone} tone",
    "interview": "Ask 3 questions about {topic} for {audience}s in a {tone} tone",
    "storytelling": "Explain {topic} as a story for {audience}s in a {tone} tone"
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style not in TEMPLATES:
        raise ValueError("Invalid style")

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=TEMPLATES[style]
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)
```

```python
# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## ✅ Task 7: Template Reusability Test

```python
reuse_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

topics = ["AI", "ML", "DL", "NLP", "Computer Vision"]

for t in topics:
    print(reuse_template.format(topic=t))
```

## 🔄 Final Pipeline Demonstration

```python
def pipeline(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    result = generate_prompt(topic, audience, tone, style)
    return result

print(pipeline("Artificial Intelligence", "beginner", "fun", "teaching"))
```
